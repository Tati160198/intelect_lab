```python
import numpy as np
from numpy import NaN
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df=pd.read_csv('nba_2013.csv')
df.head()
```


df[,sapply(df, is.numeric)]

```python
df.info()
```

```python
numeric_cols = df[[col for col, type in df.dtypes.items() if type in [np.int64, np.float64]]]
df_nan = numeric_cols.dropna(axis=1)
df_nan
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.decomposition import PCA

# Описываем модель
model = KMeans(n_clusters=5)

# Проводим моделирование
model.fit(df_nan)
```

```python
pca = PCA(n_components=2)
res = pca.fit_transform(df_nan)
plt.figure(figsize=(12,8))
plt.scatter(res[:,0], res[:,1], c=model.labels_, s=50, cmap='viridis')
plt.title('PCA')
```

```python
x_axis = df_nan['pts']  
y_axis = df_nan['ast']
plt.figure(figsize=(12,8))
plt.xlabel('Total points')
plt.ylabel('Total assists')
plt.scatter(x_axis, y_axis, c=model.labels_, s=50, cmap='viridis')
```

```python
x_axis = df_nan['age']  
y_axis = df_nan['mp']
plt.figure(figsize=(12,8))
plt.xlabel('Age')
plt.ylabel('Minutes played')
plt.scatter(x_axis, y_axis, c=model.labels_, s=50, cmap='viridis')
```
