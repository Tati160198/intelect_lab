### Pandas 


```python
import pandas as pd

data = pd.read_csv('adult.txt', sep=', ')
data.head(370)
```

#### Задача №1. Сколько в датасете мужчин, сколько женщин? (sex, male, female) ТЗ: =2 строки

```python
data['sex'].value_counts()
```

#### Задача №2. Какой средний возраст у мужчин, какой у женщин? ТЗ: =2 строки

```python
data.groupby(['sex'])['age'].mean()
```

#### Задача №3. Какой процент из общей выборки составляют представители Кубы (native-country)? ТЗ: =1 строка

```python
(data[data['native-country'] == 'Cuba'].shape[0] / data['native-country'].count())*100
```

#### Задача №4-5. Посчитайте среднее значение и стандартное отклонение возраста (age) для двух групп (salary): кто зарабатывает меньше 50K, кто зарабатывает больше 50К в год. ТЗ: =2 строки

```python
print("Средний возраст:", data.groupby(['salary'])['age'].mean())
print("Стандартное отклонение возраста:", data.groupby(['salary'])['age'].std())
```

#### Задача №6. Какой процент людей, которые зарабывают больше 50К имеют высшее образование (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)? Воспользуйтесь pd.unique() и .isin(). ТЗ: <=5 строк

```python
df_salary = data[data['salary'] == '>50K']
edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
df = df_salary[df_salary['education'].isin(edu)]
percent = (df.shape[0] / df_salary.shape[0]) * 100
print("Процент:", percent)
```

#### Задача №7. Покажите статистики по возрасту (age) для представителей разных рас (race) и гендеров (sex). Используйте groupby() и describe() методы. ТЗ: =1 строка

```python
data.groupby(['age'])['race', 'sex'].describe(percentiles=[])
```

#### Задача №8. Какие мужчины чаще зарабатывают больше 50К: женатые или одинокие (marital-status)? Примите, что женатыми являются все, у кого marital-status начинается с Married. ТЗ: <=4 строк

```python
df_salary = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]
df_salary['marital-status'].value_counts()
```

#### Задача №9. Какое максимальное значение по количеству рабочих часов в неделю (hours-per-week)? Как много людей, которые столько работают и какой процент среди них зарабатывает больше 50К? ТЗ: <=4 строк

```python
print("Максимальное количество рабочих часов в неделю: ", data['hours-per-week'].max())
df_people = data[data['hours-per-week'] == data['hours-per-week'].max()].shape[0]
df_salary = data[(data['salary'] == '>50K') & (data['hours-per-week'] == data['hours-per-week'].max())]
print("Количество людей, которые столько работают: ",df_people, "Процент зарабатывающих среди них больше 50К:", (df_salary.shape[0] / df_people) * 100)
```

```python
data.groupby(['salary','native-country'])['hours-per-week'].mean()
```
